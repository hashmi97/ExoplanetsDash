## Histogram of Discoveries by date

```python
fig = px.histogram(df, y="discoveryMethod", text_auto=True).update_yaxes(categoryorder="total ascending")
fig.show()
```

```python
fig = px.histogram(df, x="discoveryLocale", text_auto=True).update_xaxes(categoryorder="total descending")
fig.show()
```

```python
fig = px.histogram(df, x="discoveryYear", nbins=11)
fig.show()
# Create histogram of aggregates
```

```python
fig = px.histogram(df, y="discoveryFacility", text_auto=True).update_yaxes(categoryorder="total ascending")
fig.show()
```

```python
fig = px.scatter(df, x="orbitSemiMaj", y="planetMassJ", color="discoveryMethod", log_x=True, log_y=True)

fig.show()

# add another plot with the three main discovery methods that I will analyse
```
